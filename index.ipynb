### Let's do Regression 

Now, let's perform a number of simple regression experiments between the chosen independent variables and the dependent variable (price). You'll do this in a loop and in every iteration, you should pick one of the independent variables. Perform the following steps:

* Run a simple OLS regression between independent and dependent variables
* Plot a regression line on the scatter plots
* Plot the residuals using `sm.graphics.plot_regress_exog()`
* Plot a Q-Q plot for regression residuals normality test 
* Store following values in array for each iteration:
    * Independent Variable
    * r_squared'
    * intercept'
    * 'slope'
    * 'p-value'
    * 'normality (JB)' 
* Comment on each output 

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
import statsmodels.stats.api as sms


results = {"independent-var": [], "r_squared": [], "intercept": [], "slope": [], "p-value": [], "normality_JB": []}
for index, val in enumerate(new_set_list[0:len(new_set_list)-1]):
    print ("Boston Housing DataSet Regression Analysis of medv~" + val)
    print ()

    #OLS
    f = 'medv~' + val
    model = smf.ols(formula=f, data=new_set).fit()
    x_actualval = pd.DataFrame({val: [new_set[val].min(), new_set[val].max()]});
    predicted = model.predict(x_actualval)
    new_set.plot(kind='scatter', x=val, y='medv');
    plt.plot(x_actualval, predicted);
    plt.show()

    #residuals
    fig = plt.figure(figsize=(12,10))
    fig = sm.graphics.plot_regress_exog(model, val, fig=fig)
    
    #q-q plots
    fig = sm.graphics.qqplot(model.resid, dist=stats.norm, line='45', fit=True)
    plt.show()
    
    results["independent-var"].append(val)
    results["r_squared"].append(model.rsquared)
    results["intercept"].append(model.params[0])
    results["slope"].append(model.params[1])
    results["p-value"].append(model.pvalues[1])
    results["normality_JB"].append(sms.jarque_bera(model.resid)[0])
```

```python
print(results)
pd.DataFrame(results)
```
